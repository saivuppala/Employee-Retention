One cluster shows a a relationship between those with high satisfaction levels and those who recently got an evaluation. These people are likely happy and appreciated at their company. A reason for them to leave their current job might be that they were given a better opportunity and decided to take it. 

Another cluster shown in this graph is those with low satisfaction levels but recently got evaluated. These people seem to be appreciated at their company but are still unhappy. Maybe they feel that they are overqualified for their job and decided to leave. 

The last cluster of people are people with somewhat low satisfaction levels and those who have not been evaluated recently. These people probably had a good reason for leaving. They were probably not very much appreciated at work and did not like their job too much. These people probably wanted to find a better fit or possibly could have even been fired.

```python
# create correlation matrix to determine which variable are related
corrs = df.corr()


print ('correlation matrix between columns in dataframe:\n\n',corrs,'\n')

# heatmap visualization of correlation matrix
sns.heatmap(corrs, vmax=.8, square=True)
```

Looking at the different factors against the column of people who left, satisfaction level has the strongest 
correlation coefficient r value of -0.417378; a decrease in satisfaction level is correlated with an increase in the likelihood of leaving. 

After satisfaction level, time spent at the company has the third strongest r value of 0.282741. An increase in time spent at company is correlated with an increase in the likelihood of leaving. 

After time spent at company, work accident is the second strongest r value of -0.155763. However, this does not make logical sense since a decrease in work accidents seems to be correlated with the likelihood of leaving (correlation does not mean causation).

The rest of the factors (last evaluation, number of projects, average monthly hours, and promotion in the last 5 years) have weak r values and are not significant.
